```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import PillowWriter

fig=plt.figure()
l, = plt.plot([], [], 'm--')

plt.xlabel('time')
plt.ylabel('MSLALt')
plt.title('MSL')

plt.xlim(0,len(df['  AltMSL'])+100)
plt.ylim(0,max(df['  AltMSL'])+100)

def func(x):
    return (df['  AltMSL'][x])


metadata = dict(title='Movie', artist='codinglikemad')
writer = PillowWriter(fps=5, metadata=metadata)


xlist=[]
ylist=[]

with writer.saving(fig, "MSL_Alt_1.gif", 100):
    for xval in range(0,len(df['  AltMSL']),100):
        xlist.append(xval)
        ylist.append(func(xval))
        l.set_data(xlist,ylist)
        
        writer.grab_frame()
```
